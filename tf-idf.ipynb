```python
import pandas as pd
```

```python
splits = {'train': 'train.csv', 'validation': 'dev.csv', 'test': 'test.csv'}
df = pd.read_csv("hf://datasets/ibm-research/argument_quality_ranking_30k/" + splits["train"])
```

```python
df.head(3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. CrÃ©ez le vectoriseur
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)

# 2. Calculez la matrice TF-IDF sur la colonne 'argument'
tfidf_matrix = vectorizer.fit_transform(df['argument'])

# 3. Transformez en DataFrame
df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out(),
    index=df.index
)

df_tfidf.head()
```

```python
def sentence_to_tfidf(sentence):
    """
    Transforme une phrase (str) en Series TF-IDF
    basÃ©e sur le vectorizer dÃ©jÃ  entraÃ®nÃ©.
    """
    tfidf_vec = vectorizer.transform([sentence])
    return tfidf_vec.toarray()[0]
a = sentence_to_tfidf("abandon isn't keeping up with the")
```
